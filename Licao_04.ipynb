**Exercício Desafio com Visualização**

**Problema:** Um dipolo elétrico com momento de dipolo →p está inicialmente alinhado com um campo elétrico uniforme $\vec{E}$ (ângulo $\theta = \theta_0$). Se o dipolo é girado para um ângulo $\theta$ e liberado, ele oscilará.

1. Crie um gráfico da energia potencial de um dipolo elétrico em função do ângulo θ (de 0∘ a 360∘) em um campo elétrico uniforme.
2. Indique os pontos de equilíbrio estável e instável no gráfico.


```python
import numpy as np
import matplotlib.pyplot as plt

# Parâmetros do dipolo e do campo
p_magnitude = 1.0e-9  # Magnitude do momento de dipolo (C*m)
E_magnitude = 1.0e3   # Magnitude do campo elétrico uniforme (N/C)

# Ângulos de 0 a 360 graus
theta_graus = np.linspace(0, 360, 360)
theta_rad = np.radians(theta_graus) # Converter para radianos para funções trigonométricas

# Calcular a energia potencial U = -pE cos(theta)
energia_potencial = -p_magnitude * E_magnitude * np.cos(theta_rad)

# Plotar o gráfico
plt.figure(figsize=(10, 6))
plt.plot(theta_graus, energia_potencial, color='purple', linewidth=2)

# Marcar pontos de equilíbrio
# Equilíbrio estável: theta = 0 (mínimo de energia potencial)
plt.plot(0, -p_magnitude * E_magnitude, 'o', color='green', markersize=8, label='Equilíbrio Estável ($\\theta=0^\\circ$)')
plt.plot(360, -p_magnitude * E_magnitude, 'o', color='green', markersize=8) # 360 é o mesmo que 0

# Equilíbrio instável: theta = 180 (máximo de energia potencial)
plt.plot(180, p_magnitude * E_magnitude, 'o', color='red', markersize=8, label='Equilíbrio Instável ($\\theta=180^\\circ$)')

plt.title('Energia Potencial de um Dipolo Elétrico em Campo Uniforme')
plt.xlabel('Ângulo $\\theta$ (graus)')
plt.ylabel('Energia Potencial $U$ (J)')
plt.xticks(np.arange(0, 361, 45))
plt.grid(True, linestyle='--', alpha=0.7)
plt.legend()
plt.show()
```

Interpretação do Gráfico: O gráfico mostra a variação da energia potencial de um dipolo elétrico em um campo elétrico uniforme em função do ângulo θ entre o momento de dipolo $\vec{p}$ e o campo $\vec{E}$.

  - **Equilíbrio Estável ($\theta = 0^\circ$ e $θ=360^\circ$):** Nesses pontos, a energia potencial é mínima ($U=−pE$). O dipolo está alinhado com o campo elétrico, e qualquer pequena perturbação fará com que ele retorne a essa posição.
  - **Equilíbrio Instável ($\theta = 180^\circ$):** Nesse ponto, a energia potencial é máxima ($U=+pE$). O dipolo está anti-alinhado com o campo. Qualquer pequena perturbação fará com que ele se afaste dessa posição, girando para o equilíbrio estável.
  - **Comportamento Oscilatório:** Se o dipolo for liberado de um ângulo entre 0∘ e 180∘ (ou entre 180∘ e 360∘), ele experimentará um torque que o fará oscilar em torno da posição de equilíbrio estável em 0∘ (ou 360∘), de forma análoga a um pêndulo.
